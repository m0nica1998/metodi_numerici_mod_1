```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definizione della funzione di fit
def power_func(x, a, b, c):
    return a * np.power(x, b) + c

# Leggi i dati dal file
data = np.genfromtxt('/Users/monicacesario/Desktop/modulo1/susc_max_vs_L.txt')

# Estrai le colonne di interesse
x = data[:, 0]  # Dati della prima colonna
y = data[:, 1]  # Dati della seconda colonna
errors = data[:, 2]  # Errori nella terza colonna

# Effettua il fit con la funzione di fit
params, cov = curve_fit(power_func, x, y, sigma=errors)

# Estrai i parametri del fit e i relativi errori
a, b, c = params
a_err, b_err, c_err = np.sqrt(np.diag(cov))

# Calcola il chi quadro ridotto
residuals = y - power_func(x, a, b, c)
reduced_chi_square = np.sum((residuals / errors) ** 2) / (len(x) - 3)

# Calcola i residui normalizzati
normalized_residuals = residuals / errors

# Crea il grafico dei punti con la curva di fit sovrapposta
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.errorbar(x, y, yerr=errors, fmt='.', color='black', capsize=2)
plt.plot(x, power_func(x, a, b, c), color='#006699')
plt.xlabel("L")
plt.ylabel("χₘₐₓ")

plt.subplot(1, 2, 2)
plt.plot(x, normalized_residuals, '.', color='black')
plt.axhline(0, color='red', linewidth=0.5)
plt.xlabel("L")
plt.ylabel("Normalized Residuals")

plt.tight_layout()
plt.show()

# Stampa i parametri del fit con i relativi errori e il chi quadro ridotto
print("Parametri del fit:")
print("a =", a, "+/-", a_err)
print("b =", b, "+/-", b_err)
print("c =", c, "+/-", c_err)
print("Chi quadro ridotto =", reduced_chi_square)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Funzione di fit
def fit_function(x, a, b):
    return (b - np.sinh(2*x)**(-4))**a

# Leggi i dati dal file
data = np.genfromtxt('/Users/monicacesario/Desktop/modulo1/cold_magn40.dat')

# Estrai le colonne di interesse
x = data[:, 0]  # Dati della prima colonna traslati
y = data[:, 1]  # Dati della seconda colonna
errors = data[:, 2]  # Errori nella terza colonna

# Escludi i primi due punti
x = x[1:]
y = y[1:]
errors = errors[1:]

# Effettua il fit dei dati
params, cov = curve_fit(fit_function, x, y, sigma=errors)

# Estrai i parametri del fit e i relativi errori
a, b = params
a_err, b_err = np.sqrt(np.diag(cov))

# Calcola il chi quadro ridotto
residuals = y - fit_function(x, a, b)
reduced_chi_square = np.sum((residuals / errors) ** 2) / (len(x) - 2)

# Crea il grafico dei punti con gli errori
plt.errorbar(x, y, yerr=errors, fmt='.', color='black', capsize=2)
plt.plot(x, fit_function(x, a, b), color='#006699')
plt.xlabel(r'$\beta$')
plt.ylabel("|M|")

# Mostra il grafico
plt.show()

# Stampa il valore del parametro a, l'errore e il chi quadro ridotto
print("Parametri del fit:")
print("a =", a, "+/-", a_err)
print("b =", b, "+/-", b_err)
#print("c =", c, "+/-", c_err)
print("Chi quadro ridotto =", reduced_chi_square)

```
