```python
import matplotlib.pyplot as plt

# Definisci i nomi del file di input
nome_file = "/Users/monicacesario/Desktop/obs1_vs_k.dat"

# Leggi i dati dal file
x = []
y = []
with open(nome_file, "r") as file:
    for riga in file:
        dati = riga.split()
        x.append(float(dati[0]))
        y.append(float(dati[1]))

# Crea il grafico
plt.plot(x, y, 'o-')


plt.grid(True)

# Mostra il grafico
plt.show()


```
