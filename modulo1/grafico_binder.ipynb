```python
import numpy as np
import matplotlib.pyplot as plt

# Leggi i dati dal file
data = np.loadtxt('/Users/monicacesario/Desktop/Binder_035.dat')

# Impostazione font
font1 = {'family': 'Times New Roman', 'size': 15}
font2 = {'family': 'Times New Roman', 'size': 12}

x = data[:, 0]  # Prima colonna
y = data[:, 1]  # Seconda colonna
err = data[:, 2]  # Terza colonna (errori)

# Definisci la funzione di fit (parabola)
def fit_func(x, a, b, c):
    return a * x ** 2 + b * x + c

# Esegui il fit dei dati
params, cov_matrix = np.polyfit(1 / x, y, deg=2, w=1 / err, cov=True)

# Estrai i parametri del fit e le relative incertezze
fit_params = params
fit_errors = np.sqrt(np.diag(cov_matrix))

# Calcola il chi-quadro ridotto
residuals = y - fit_func(1 / x, *fit_params)
reduced_chi_squared = np.sum((residuals / err) ** 2) / (len(x) - len(fit_params))

# Crea il vettore di x per il grafico della funzione di fit
x_fit = np.linspace(min(x), max(x), 100)
y_fit = fit_func(1 / x_fit, *fit_params)

# Crea il grafico dei punti dati con la funzione di fit sovrapposta
plt.errorbar(1 / x, y, yerr=err, fmt='k.', ecolor='black', capsize=2)
plt.plot(1 / x_fit, y_fit, '#006699')

plt.title('Cumulante di Binder per $\\beta$ = 0.35', fontdict=font1)
plt.xlabel('1 / L', fontdict=font2)
plt.ylabel('C.B.', fontdict=font2)

# Aggiungi le tacchette piccole sugli assi
plt.minorticks_on()
plt.tick_params(which='both', direction='in', top=True, right=True)

# Stampa a schermo i risultati del fit
print("Parametri del fit:")
param_names = ['a', 'b', 'c']
for i, param in enumerate(fit_params):
    print(f"{param_names[i]}: {param:.4f} +/- {fit_errors[i]:.4f}")

print(f"Chi-quadro ridotto: {reduced_chi_squared:.4f}")

# Mostra il grafico a schermo

plt.show()
```
